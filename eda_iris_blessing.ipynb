## 4. Descriptive Statistics

Summary statistics show that the average sepal length is about 5.84 cm with a range from 4.3 to 7.9 cm. Petal lengths have more variability, ranging from 1.0 to 6.9 cm.

The dataset appears well-balanced without missing values, as shown by the count of 150 for all features.


```python

## Visualize Data Distributions
iris_df.hist()
plt.show()
for col in iris_df.select_dtypes(include=['object', 'category']).columns:
    sns.countplot(x=col, data=iris_df)
    plt.title(f'Distribution of {col}')
    plt.show()
iris_df.rename(columns={'sepal_length': 'Sepal Length'}, inplace=True)
iris_df['Sepal Area'] = iris_df['Sepal Length'] * iris_df['sepal_width']
iris_df.head()
## Visualize Feature Relationships
sns.pairplot(iris_df, hue='species')
plt.show()
sns.scatterplot(data=iris_df, x='Sepal Length', y='Sepal Area', hue='species')
plt.show()

```

## Initial Data Inspection

The Iris dataset contains 150 rows and 5 columns. The columns are numerical measurements of iris flowers (sepal length, sepal width, petal length, petal width) and one categorical column, `species`, which indicates the iris species (setosa, versicolor, virginica).

The data types are appropriate with numerical columns as floats and species as categorical.


# Summary of Exploratory Data Analysis on Iris Dataset

- The Iris dataset contains **150 samples** with **5 columns**: sepal length, sepal width, petal length, petal width, and species.
- The data has **3 species classes**: Setosa, Versicolor, and Virginica.
- Summary statistics show that:
  - Setosa has smaller sepal and petal dimensions on average.
  - Virginica generally has the largest measurements.
- Visualizations revealed:
  - Clear separation between species in petal length and width.
  - Sepal length correlates positively with sepal area.
  - The species Setosa is distinctly different from the other two in most features.
- No missing data was detected.
- These insights can guide classification and predictive modeling tasks.
